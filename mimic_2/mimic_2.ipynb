```python
import ehrapy as ep
```

```python
adata = ep.dt.mimic_2(encoded=True)
```

```python
ep.pp.knn_impute(adata, n_neighbours=20)
```

```python
ep.pp.log_norm(adata, vars=["iv_day_1"], offset=1)
```

```python
ep.pp.pca(adata)
```

```python
ep.pp.neighbors(adata)
```

```python
ep.tl.umap(adata)
```

```python
ep.pl.umap(
    adata,
    color=["day_icu_intime", "service_unit"],
    wspace=0.5,
    title=["Day of ICU admission", "Service unit"],
)
```

```python
ep.io.write("mimic_2_preprocessed.h5ad", adata)
```

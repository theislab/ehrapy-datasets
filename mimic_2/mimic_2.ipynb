```python
import ehrapy as ep
import ehrdata as ed
```

```python
adata = ep.dt.mimic_2(encoded=True)
edata = ed.EHRData.from_adata(adata=adata)
```

```python
ep.pp.knn_impute(edata, n_neighbours=20)
```

```python
ep.pp.log_norm(edata, vars=["iv_day_1"], offset=1)
```

```python
ep.pp.pca(edata)
```

```python
ep.pp.neighbors(edata)
```

```python
ep.tl.umap(edata)
```

```python
ep.pl.umap(
    edata,
    color=["day_icu_intime", "service_unit"],
    wspace=0.5,
    title=["Day of ICU admission", "Service unit"],
)
```

```python
ep.io.write("mimic_2_preprocessed.h5ad", edata)
```
